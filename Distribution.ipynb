# Probability Distribution

### 01.Binomial Distribution

```python
from scipy.stats import binom
import numpy as np


def binomial(n,p,k):
    return binom.pmf(k,n,p)

n=int(input("Enter the number of trials: "))
p=float(input("Enter the probability of success: "))
k=int(input("Enter the number of successes: "))

print(binomial(n,p,k)) 

import matplotlib.pyplot as plt

plt.plot(range(n+1),binomial(n,p,range(n+1)))
plt.show()
```

### 02.Normal Distribution

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

probs=[norm.pdf(x,loc=0,scale=1) for x in np.linspace(-5,5,100)]
plt.plot(np.linspace(-5,5,100),probs)
plt.show()
```
